```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("cleaned.csv")
```

```python
df
```

```python
# Step 1: Count how many times each payment type appears
payment_counts = df['payment_type'].value_counts()

# Step 2: Plot a pie chart from those counts
plt.pie(
    payment_counts,                      # The values (how big each slice is)
    labels=payment_counts.index,         # The names for each slice
    autopct='%1.1f%%',                   # Show percentages with 1 decimal place
    startangle=140                       # Rotate it a bit for visual balance
)

plt.title('Payment Type Distribution')
#plt.axis('equal')  # This keeps the pie circle-shaped (not an oval)
plt.legend()
plt.show()
```

```python
reviews=df['review_category'].value_counts()
plt.bar(reviews.index,reviews)
plt.title("review distribution")
```
